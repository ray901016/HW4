```python
import pandas as pd
from pycaret.classification import *

# 下載 Titanic 資料集
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

# 預處理數據
data = data[['Survived', 'Pclass', 'Sex', 'Age', 'Fare', 'Embarked']]  # 篩選需要的欄位
data['Age'] = data['Age'].fillna(data['Age'].median())  # 填補 Age 缺失值
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])  # 填補 Embarked 缺失值
data = pd.get_dummies(data, columns=['Sex', 'Embarked'], drop_first=True)  # 類別變數轉換

# 初始化 PyCaret 環境
clf = setup(data=data, target='Survived', session_id=123, verbose=False)

# 比較所有模型
best_model = compare_models()

# 打印最佳模型
print("Best Model:", best_model)
```

```python
# 評估最佳模型
evaluate_model(best_model)
# 保存最佳模型
save_model(best_model, 'best_titanic_model')

# 將模型下載到本地
from google.colab import files
files.download('best_titanic_model.pkl')
```
